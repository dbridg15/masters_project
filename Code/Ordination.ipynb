```python
# # covariance matrix
cov_mat = np.cov(x.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

# correlation matrix can also be used...
cor_mat1 = np.corrcoef(x.T)
cor_mat2 = np.corrcoef(x.T)

eig_vals1, eig_vecs1 = np.linalg.eig(cor_mat1)
eig_vals2, eig_vecs2 = np.linalg.eig(cor_mat2)

# eigenvals and eigenvecs for all are the same...

# u is eigvectors, 
u,s,v = np.linalg.svd(x.T)
```

```python
for ev in eig_vecs:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print("Its all good!")
```

```python
# Make a list of (eigenvalue, eigenvector) tuples# Make a 
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
```

```python
eig_pairs.sort()
eig_pairs.reverse()
```

```python
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
tot = sum(eig_vals)
var_exp = [(i / tot) for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```

```python
matrix_w = np.hstack((eig_pairs[0][1].reshape(4,1), 
                      eig_pairs[1][1].reshape(4,1)))
```

```python
Y = x_std.dot(matrix_w)

plot_df = pd.DataFrame({'name'   : [],
                        'x_vals' : [],
                        'y_vals' : []})

for name in names:
    x_vals = Y[y == name, 0]
    y_vals = Y[y == name, 1]
    tmp    = pd.DataFrame({'name'   : name,
                          'x_vals' : x_vals,
                          'y_vals' : y_vals})
    
    frames  = [plot_df, tmp]
    plot_df = pd.concat(frames)
```

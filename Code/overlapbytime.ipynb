```r
require("dplyr")
```

```r
trees_df = read.csv("../Results/trees_sorted.csv")
trees_df$date = as.Date(trees_df$date)
```

```r
plots      = sort(rep(as.character(unique(trees_df$plot)), 4))
census     = rep(seq(1,4), length(unique(trees_df$plot)))
test       = as.data.frame(cbind(plots, census))
test$start = NA
test$end   = NA
```

```r
for(i in 1:nrow(test)){
    tmp = subset(trees_df, plot == test[i, "plots"] & census == test[i, "census"])
    if (nrow(tmp) != 0 ){
        test$start[i] = min(tmp$date, na.rm = T)
        test$end[i]   = max(tmp$date, na.rm = T)
    }
}
test$middle = (test$start+test$end)/2
```

```r
test
```

```r
plots = sort(rep(as.character(unique(trees_df$plot)), 3))
steps = rep(c("1-2", "2-3", "3-4"), length(unique(trees_df$plot)))

test2 = as.data.frame(cbind(plots, steps))
```

```r
test2$endtostart = NA
test2$midtomid   = NA
```

```r
for( i in 1:nrow(test2)){
    a = as.character(test2$plot[i])
    b = as.integer(unlist(strsplit(as.character(test2$steps[i]), "-"))[1])
    c = b + 1

    end1   = test$end[which(test$plot == a & test$census == b)]
    start2 = test$start[which(test$plot == a & test$census == c)]

    mid1 = test$middle[which(test$plot == a & test$census == b)]
    mid2 = test$middle[which(test$plot == a & test$census == c)]

    test2$endtostart[i] = (start2 - end1)/365
    test2$midtomid[i] = (mid2 - mid1)/365
}
    
```
